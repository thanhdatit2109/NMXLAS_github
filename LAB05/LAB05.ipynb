```python
import cv2
import numpy as np

# Đọc ảnh
img1 = cv2.imread(r"C:\Users\dat21\exercise\bird.png", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(r"C:\Users\dat21\exercise\dalat.jpg", cv2.IMREAD_GRAYSCALE)

# Kiểm tra ảnh đã load chưa
if img1 is None:
    print("Ảnh 1 không tồn tại hoặc sai đường dẫn!")
    exit()

if img2 is None:
    print("Ảnh 2 không tồn tại hoặc sai đường dẫn!")
    exit()

# Hàm tìm điểm góc
def tim_diem_goc(img):
    img_float = np.float32(img)
    dst = cv2.cornerHarris(img_float, blockSize=2, ksize=3, k=0.04)
    dst = cv2.dilate(dst, None)
    return dst

# Gọi hàm
harris1 = tim_diem_goc(img1)
harris2 = tim_diem_goc(img2)

print("Harris1 shape:", harris1.shape)
def tao_keypoint(dst):
    keypoints = []
    y, x = np.where(dst > 0.01 * dst.max())
    for i in range(len(x)):
        # Ép kiểu float() để đảm bảo đúng tham số
        kp = cv2.KeyPoint(float(x[i]), float(y[i]), 20)
        keypoints.append(kp)
    return keypoints


kp1 = tao_keypoint(harris1)
kp2 = tao_keypoint(harris2)

# Tạo đối tượng ORB
orb = cv2.ORB_create()

# Tính descriptor cho các điểm góc
kp1, des1 = orb.compute(img1, kp1)
kp2, des2 = orb.compute(img2, kp2)

# So khớp bằng BFMatcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)

# Sắp xếp theo khoảng cách
matches = sorted(matches, key=lambda x: x.distance)

# Vẽ 30 cặp điểm khớp tốt nhất
anh_ket_qua = cv2.drawMatches(img1, kp1, img2, kp2, matches[:30], None, flags=2)

# Hiển thị ảnh kết quả
cv2.imshow('Ket qua so khop', anh_ket_qua)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
