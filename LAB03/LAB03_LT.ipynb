```python
#BT1
from PIL import Image
import numpy as np
from scipy.ndimage import map_coordinates


kiwi = Image.open("kiwi.jpegS")

kiwi = kiwi.convert("RGB")


translated = Image.new("RGB", (kiwi.width + 50, kiwi.height + 30))
translated.paste(kiwi, (50, 30))


arr = np.array(translated)
coords = np.meshgrid(np.arange(arr.shape[0]), np.arange(arr.shape[1]), indexing='ij')
coords_wave = np.array([
    coords[0] + 10 * np.sin(coords[1] / 20.0),
    coords[1] + 10 * np.sin(coords[0] / 20.0)
])
warped = np.zeros_like(arr)
for i in range(3): 
    warped[..., i] = map_coordinates(arr[..., i], coords_wave, order=1, mode='reflect')

Image.fromarray(warped).save("kiwi_wave.jpg")
```

```python
#BT3
from PIL import Image, ImageOps


mountain = Image.open("mountain.jpeg").convert("RGBA")
boat = Image.open("boat.jpeg").convert("RGBA")


mountain_rot = mountain.rotate(45, expand=False)
boat_rot = boat.rotate(45, expand=False)


mountain_mirror = ImageOps.flip(mountain_rot)
boat_mirror = ImageOps.flip(boat_rot)


canvas = Image.new("RGB", (mountain_mirror.width + boat_mirror.width, max(mountain_mirror.height, boat_mirror.height)), "white")
canvas.paste(mountain_mirror.convert("RGB"), (0, 0))
canvas.paste(boat_mirror.convert("RGB"), (mountain_mirror.width, 0))
canvas.save("mountain_boat_mirror.jpg")
```

```python
from PIL import Image, ImageFilter
import numpy as np
from scipy.ndimage import map_coordinates

def transform_image(img):
    while True:
        print("\n==== MENU BI·∫æN ƒê·ªîI ====")
        print("1. T·ªãnh ti·∫øn ·∫£nh")
        print("2. Xoay ·∫£nh")
        print("3. Ph√≥ng to / Thu nh·ªè ·∫£nh")
        print("4. L√†m m·ªù Gaussian")
        print("5. Bi·∫øn d·∫°ng s√≥ng")
        print("0. Tho√°t v√† l∆∞u k·∫øt qu·∫£")
        choice = input("üëâ Nh·∫≠p l·ª±a ch·ªçn c·ªßa b·∫°n (0-5): ")

        if choice == "1":
            dx = int(input("Nh·∫≠p s·ªë pixel d·ªãch sang ph·∫£i: "))
            dy = int(input("Nh·∫≠p s·ªë pixel d·ªãch xu·ªëng: "))
            new_img = Image.new("RGB", (img.width + dx, img.height + dy), "white")
            new_img.paste(img, (dx, dy))
            img = new_img

        elif choice == "2":
            angle = float(input("Nh·∫≠p g√≥c xoay (ƒë·ªô): "))
            reshape = input("Reshape = True (co gi√£n ·∫£nh)? (y/n): ").lower() == "y"
            img = img.rotate(angle, expand=reshape)

        elif choice == "3":
            zoom = float(input("Nh·∫≠p h·ªá s·ªë zoom (VD: 0.5 ho·∫∑c 2.0): "))
            new_size = (int(img.width * zoom), int(img.height * zoom))
            img = img.resize(new_size)

        elif choice == "4":
            sigma = float(input("Nh·∫≠p ƒë·ªô m·ªù (sigma): "))
            img = img.filter(ImageFilter.GaussianBlur(radius=sigma))

        elif choice == "5":
            amplitude = float(input("Nh·∫≠p bi√™n ƒë·ªô s√≥ng: "))
            arr = np.array(img)
            coords = np.meshgrid(np.arange(arr.shape[0]), np.arange(arr.shape[1]), indexing='ij')
            coords_wave = np.array([
                coords[0] + amplitude * np.sin(coords[1] / 20.0),
                coords[1] + amplitude * np.sin(coords[0] / 20.0)
            ])
            warped = np.zeros_like(arr)
            for i in range(3):  
                warped[..., i] = map_coordinates(arr[..., i], coords_wave, order=1, mode='reflect')
            img = Image.fromarray(warped)

        elif choice == "0":
            break

        else:
            print(" L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá. Vui l√≤ng nh·∫≠p l·∫°i.")

    img.save("transformed_image.jpg")
    print(" ƒê√£ l∆∞u k·∫øt qu·∫£ th√†nh transformed_image.jpg")

# --- KH·ªûI ƒê·ªòNG ---
print(" B√ÄI 5: Ch∆∞∆°ng tr√¨nh menu bi·∫øn ƒë·ªïi ·∫£nh")
img_path = input(" Nh·∫≠p ƒë∆∞·ªùng d·∫´n ·∫£nh (VD: C:/Users/dat21/New folder (2)/kiwi.jpg): ")
try:
    img = Image.open(img_path).convert("RGB")
    transform_image(img)
except FileNotFoundError:
    print(" Kh√¥ng t√¨m th·∫•y ·∫£nh. Vui l√≤ng ki·ªÉm tra l·∫°i ƒë∆∞·ªùng d·∫´n.")
```
