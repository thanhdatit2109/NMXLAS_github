## C√¢u 1 (2 ƒëi·ªÉm) Ch·ªçn m·ªôt ·∫£nh b·∫•t k·ª≥ (t·ª± ƒë·∫∑t t√™n, v√≠ d·ª• `pic1.jpg`) v√† th·ª±c hi·ªán:

- L√†m m·ªù ·∫£nh b·∫±ng box filter. (0.5 ƒëi·ªÉm)  
- √Åp d·ª•ng Laplacian ƒë·ªÉ ph√°t hi·ªán bi√™n. (0.5 ƒëi·ªÉm)  
- Chuy·ªÉn ƒë·ªïi ·∫£nh sang ·∫£nh √¢m b·∫£n (negative). (0.5 ƒëi·ªÉm)  
- Chuy·ªÉn sang kh√¥ng gian m√†u HSV v√† l∆∞u 3 k√™nh H, S, V ri√™ng bi·ªát.  l∆∞u th√†nh ·∫£nh grayscale t∆∞∆°ng ·ª©ng (`[ten_anh]_L.jpg`, `[ten_anh]_A.jpg`, `[ten_anh]_B.jpg`). (0.5 ƒëi·ªÉm) (0.5 ƒëi·ªÉm)


```python
import cv2
import numpy as np

img = cv2.imread('pic1.jpg')

blur = cv2.blur(img, (5, 5))
cv2.imwrite('pic1_blur.jpg', blur)
print('ƒê√£ l√†m m·ªù ·∫£nh')


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(gray, cv2.CV_64F)
lap = cv2.convertScaleAbs(lap)
cv2.imwrite('pic1_laplacian.jpg', lap)
print(' ƒê√£ √°p d·ª•ng Laplacian')


negative = 255 - img
cv2.imwrite('pic1_negative.jpg', negative)
print(' ƒê√£ chuy·ªÉn sang ·∫£nh √¢m b·∫£n')


hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
cv2.imwrite('pic1_H.jpg', h)
cv2.imwrite('pic1_S.jpg', s)
cv2.imwrite('pic1_V.jpg', v)
print(' ƒê√£ l∆∞u k√™nh H, S, V c·ªßa HSV')
```

### C√¢u 2 (4 ƒëi·ªÉm) Vi·∫øt m·ªôt ch∆∞∆°ng tr√¨nh Python s·ª≠ d·ª•ng OpenCV ƒë·ªÉ t·∫°o menu t∆∞∆°ng t√°c cho ph√©p ng∆∞·ªùi d√πng ch·ªçn c√°c k·ªπ thu·∫≠t chuy·ªÉn ƒë·ªïi kh√¥ng gian m√†u v√† x·ª≠ l√Ω ·∫£nh n√¢ng cao t·ª´ m·ªôt danh s√°ch, √°p d·ª•ng ƒë·ªìng th·ªùi cho nhi·ªÅu ·∫£nh.

### Y√™u c·∫ßu:

1. Menu g·ªìm:  
* Chuy·ªÉn sang ·∫£nh x√°m (Grayscale) (0.5 ƒëi·ªÉm)  
* Chuy·ªÉn sang HSV (0.5 ƒëi·ªÉm)  
* Chuy·ªÉn sang LAB (0.5 ƒëi·ªÉm)  
* C√¢n b·∫±ng histogram (0.5 ƒëi·ªÉm)  
* Adaptive Thresholding (tham s·ªë ng·∫´u nhi√™n) (0.5 ƒëi·ªÉm)  
* CLAHE (Contrast Limited Adaptive Histogram Equalization) (0.5 ƒëi·ªÉm)

2. Ch∆∞∆°ng tr√¨nh x·ª≠ l√Ω ƒë·ªìng th·ªùi 3 ·∫£nh b·∫•t k·ª≥ do sinh vi√™n t·ª± ch·ªçn (c√≥ th·ªÉ ch·ªçn b·∫±ng ƒë∆∞·ªùng d·∫´n file ho·∫∑c nh·∫≠p t√™n ·∫£nh t√πy √Ω). (0.5 ƒëi·ªÉm)

3. Ph√≠m t∆∞∆°ng ·ª©ng ƒë·ªÉ k√≠ch ho·∫°t c√°c ph∆∞∆°ng ph√°p x·ª≠ l√Ω:  
* G: Grayscale  
* H: HSV  
* L: LAB  
* Q: Histogram Equalization  
* A: Adaptive Threshold  
* C: CLAHE (0.5 ƒëi·ªÉm)

4. L∆∞u file k·∫øt qu·∫£ v·ªõi ƒë·ªãnh d·∫°ng: `result_[ph∆∞∆°ng ph√°p]_[t√™n ·∫£nh g·ªëc].jpg`  
   V√≠ d·ª•: `result_gray_flower.jpg`, `result_clahe_img1.jpg` (0.5 ƒëi·ªÉm)


```python
import cv2
import numpy as np
import random

img_names = ['img1.jpg', 'img2.jpg', 'img3.jpg']  # üëâ ƒê·ªïi theo t√™n ·∫£nh th·∫≠t c·ªßa b·∫°n


imgs = []
img_names_valid = []
for name in img_names:
    img = cv2.imread(name)
    if img is not None:
        imgs.append(img)
        img_names_valid.append(name)
    else:
        print(f"‚ùå Kh√¥ng ƒë·ªçc ƒë∆∞·ª£c ·∫£nh: {name}")

# --- H√†m l∆∞u k·∫øt qu·∫£ ---
def save_result(imgs, method):
    for i, img in enumerate(imgs):
        base_name = img_names_valid[i].split('.')[0]
        out_name = f'result_{method}_{base_name}.jpg'
        cv2.imwrite(out_name, img)
        print(f'‚úÖ ƒê√£ l∆∞u: {out_name}')

# --- Hi·ªÉn th·ªã menu ---
print("====== MENU X·ª¨ L√ù ·∫¢NH (Jupyter version) ======")
print("G - Grayscale")
print("H - HSV")
print("L - LAB")
print("Q - Histogram Equalization")
print("A - Adaptive Threshold")
print("C - CLAHE")
print("Esc - Tho√°t ch∆∞∆°ng tr√¨nh")



while True:
    key = input("\n Nh·∫≠p ph√≠m x·ª≠ l√Ω (G/H/L/Q/A/C ho·∫∑c esc ƒë·ªÉ tho√°t): ").lower()

    if key == 'esc':
        print(" K·∫øt th√∫c ch∆∞∆°ng tr√¨nh.")
        break 

    result = []
    method = None

    for img in imgs:
        if key == 'g':
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            result.append(gray)
            method = 'gray'
        elif key == 'h':
            hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
            result.append(hsv)
            method = 'hsv'
        elif key == 'l':
            lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
            result.append(lab)
            method = 'lab'
        elif key == 'q':
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            eq = cv2.equalizeHist(gray)
            result.append(eq)
            method = 'equalized'
        elif key == 'a':
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            blockSize = random.choice([11, 13, 15])
            C = random.randint(2, 10)
            thres = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, blockSize, C)
            result.append(thres)
            method = 'adaptive'
        elif key == 'c':
            lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
            l, a, b = cv2.split(lab)
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
            cl = clahe.apply(l)
            merged = cv2.merge((cl, a, b))
            clahe_bgr = cv2.cvtColor(merged, cv2.COLOR_LAB2BGR)
            result.append(clahe_bgr)
            method = 'clahe'

    if method is not None:
        save_result(result, method)
    else:
        print(" Ph√≠m kh√¥ng h·ª£p l·ªá. Vui l√≤ng nh·∫≠p G, H, L, Q, A ho·∫∑c C.")

```

### C√¢u 3 (4 ƒëi·ªÉm) Vi·∫øt m·ªôt ch∆∞∆°ng tr√¨nh Python ƒë·ªÉ x·ª≠ l√Ω 3 ·∫£nh b·∫•t k·ª≥ do sinh vi√™n t·ª± ch·ªçn.

* C·∫Øt ·∫£nh ƒë·∫ßu ti√™n theo t·ªâ l·ªá 80% ·ªü gi·ªØa. (0.5 ƒëi·ªÉm)  
* Xoay ·∫£nh th·ª© hai 90 ƒë·ªô v√† l·∫≠t d·ªçc. (0.5 ƒëi·ªÉm)  
* Thu nh·ªè ·∫£nh th·ª© ba xu·ªëng 1/3 k√≠ch th∆∞·ªõc ban ƒë·∫ßu v√† √°p d·ª•ng Median Blur v·ªõi kernel 7x7. (1.5 ƒëi·ªÉm)  
* Thay ƒë·ªïi ƒë·ªô s√°ng v√† ƒë·ªô t∆∞∆°ng ph·∫£n ·∫£nh th·ª© ba theo c√¥ng th·ª©c:

$$
I_{out}(x, y) = \alpha \cdot I_{in}(x, y) + \beta
$$

Trong ƒë√≥:  

$$
\alpha \in [0.7, 1.8], \quad \beta \in [-40, 40]
$$

Gi√° tr·ªã ƒë·∫ßu ra c·∫ßn ƒë∆∞·ª£c gi·ªõi h·∫°n trong kho·∫£ng [0, 255] b·∫±ng c√¥ng th·ª©c:

$$
I_{out}(x, y) = \text{clip}(I_{out}(x, y), 0, 255)
$$


```python
import cv2
import numpy as np
import random

img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')
img3 = cv2.imread('img3.jpg')

h, w = img1.shape[:2]
x1, y1 = int(0.1 * w), int(0.1 * h)
x2, y2 = int(0.9 * w), int(0.9 * h)
cropped = img1[y1:y2, x1:x2]
cv2.imwrite('crop_img1.jpg', cropped)
print('C·∫Øt ·∫£nh gi·ªØa (80%)')


rotated = cv2.rotate(img2, cv2.ROTATE_90_CLOCKWISE)
flipped = cv2.flip(rotated, 0)
cv2.imwrite('flip_rotate_img2.jpg', flipped)
print(' Xoay v√† l·∫≠t ·∫£nh 2')


resized = cv2.resize(img3, (img3.shape[1]//3, img3.shape[0]//3))
blurred = cv2.medianBlur(resized, 7)
cv2.imwrite('resize_blur_img3.jpg', blurred)
print('Thu nh·ªè + l√†m m·ªãn ·∫£nh 3')

    
alpha = random.uniform(0.7, 1.8)
beta = random.randint(-40, 40)
adjusted = np.clip(alpha * blurred + beta, 0, 255).astype(np.uint8)
cv2.imwrite('bright_contrast_img3.jpg', adjusted)
print(f' ƒêi·ªÅu ch·ªânh ƒë·ªô s√°ng (alpha={alpha:.2f}) v√† beta={beta}')
```
